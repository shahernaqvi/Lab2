```python
import matplotlib.pyplot as plt
import numpy as np

# Open the file in read mode
file = open("genome.txt", "r")
# Read the entire content of the file
content = file.read()
# Print the content
print(content)

# Close the file
file.close()
nucleotides=list(content)
genome_length=len(nucleotides)
print(genome_length)
# We'll use the parametric equations for a helix:
# x = cos(t), y = sin(t), z = t (or a scaled version of t)
# We want to span a range so that the helix makes a few turns.
t = np.linspace(0, 4 * np.pi, genome_length) # 4*pi gives about

x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length) # z increases linearly to

# Combine the coordinates into a (genome_length x 3) array
coordinates = np.column_stack((x, y, z))
# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# Color mapping
colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
for i, nucleotide in enumerate(nucleotides):
  ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],
  color=colors[nucleotide], marker='o')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = np.array(iris.data)  # Features
Y = np.array(iris.target)  # Target variable

num = X.shape[1]
for i in range(num):
    data = X[:, i]

    mean = np.mean(data)
    median = np.median(data)
    stddev = np.std(data)
    minval = np.min(data)
    maxval = np.max(data)

    print(f"Feature {i+1}:")
    print(f"  Mean: {mean:.2f}")
    print(f"  Median: {median:.2f}")
    print(f"  Standard Deviation: {stddev:.2f}")
    print(f"  Minimum: {minval:.2f}")
    print(f"  Maximum: {maxval:.2f}")

sepal_data = X[:, :2]
print("Sepal Length and Width Data:\n", sepal_data)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Sepal Width")
plt.colorbar(label="Species", ticks=[0, 1, 2])
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=20, color='red', edgecolor='black')  # Adjust bins for better visualization
plt.xlabel("Length")
plt.ylabel("F")
plt.title("Distribution of Sepal Length")
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(X[:, 2], X[:, 3], marker='o', linestyle='-', color='blue')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs Petal Width")
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

try:
    img = Image.open("cats.jpg")
    numpy_array = np.array(img)
    print(numpy_array.shape)

    # Plot the Original Image
    plt.imshow(numpy_array)
    plt.axis('off')
    plt.title('Original Image')
    plt.show()

    # Rotate the Image
    rotated_array = np.rot90(numpy_array)
    plt.imshow(rotated_array)
    plt.axis('off')
    plt.title('Rotated Image')
    plt.show()

    # Flip the Image Horizontally
    flipped_array = np.fliplr(numpy_array)
    plt.imshow(flipped_array)
    plt.axis('off')
    plt.title('Flipped Image')
    plt.show()

    # Apply Grayscale Filter
    def grayscale(rgb):
        return np.dot(rgb[..., :3], [0.2989, 0.5870, 0.1140])

    gray_array = grayscale(numpy_array)

    plt.imshow(gray_array, cmap=plt.get_cmap('gray'))  # Use a grayscale colormap
    plt.axis('off')
    plt.title('Grayscale Image')
    plt.show()

except FileNotFoundError:
    print("Error: The image file 'pic.jpg' was not found.  Make sure it's in the same directory as your script, or provide the correct path.")
except Exception as e:
    print(f"An error occurred: {e}")
```
